Movie App

```python
def is_exist(movies, movie):
  if(movie in movies):
    return True
  else:
    return False

def display(movies):
  print(f"{len(movies)} movies in total: ")
  for key, value in movies.items():
    print(f"The movie {key}"+f" has rating {value}")

def create(movies, movie,rating):
  movies[movie] = rating
  print("The Movie added Successfully.")
  print(movies)

def read(movies, movie):
  if(movie in movies):
    print(f"The movie {movie}"+f" has rating {movies.get(movie)}")
  else:
    print("This movie does not exit.")

def update(movies, movie, rating):
  movies[movie] = rating
  print(f"The movie {movie}"+f" has now rating {rating}")

def delete(movies, movie):
  del movies[movie]
  print(f"The movie {movie}"+f" has been deleted from the dictionary.")

def top_rated_movies(movies):
  print("Top rated Movies in the dictionary: ")

  for key, value in movies.items():
    if(value > 9.0):
      print(f"\tThe movie {key}"+f" has rating {value}")

def least_rated_movies(movies):
  print("Least Rated Movies in the dictionary: ")

  for k, v in movies.items():
    if(v < 4.0):
      print(f"\tThe movie {k}"+f" has rating {v}")

def sort(movies):
  print("The movies has been successfully sorted by rating: ")

  sorted_movies = dict(sorted(movies.items(), key=lambda item: item[1]))
  print(sorted_movies)

def rand(movies):
  import random
  if movies:
    ran = random.choice(list(movies.keys()))
    print(f"The randomly selected movie is {ran} with rating {movies[ran]}")
  else:
    print("No movies available to select.")

def statistics(movies):
  avg_rating = sum(movies.values())/len(movies)
  best_mov = max(movies, key = movies.get)
  worst_mov = min(movies, key = movies.get)

  sort_movies = sorted(movies.values())
  if len(sort_movies) % 2 == 0:
    mid1 = len(sort_movies) // 2 - 1
    mid2 = mid1 + 1
    median = (sort_movies[mid1] + sort_movies[mid2]) / 2
  else:
    mid = len(sort_movies) // 2
    median = sort_movies[mid]


  print(f"Average Rating: {avg_rating}")
  print(f"Median Rating: {median}")
  print(f"Best Movie: {best_mov}")
  print(f"Worst Movie: {worst_mov}")

def analytics(movies):
  analoop =True
  while analoop:
    print("Press 0 to go back")
    print("Press 1 to display top rated movies.")
    print("Press 2 display least rated movies.")
    ana = int(input("Enter your choice: "))

    if(ana == 0):
      analoop = False

    elif(ana == 1):
      top_rated_movies(movies)

    elif(ana == 2):
      least_rated_movies(movies)

    else:
      print("Invalid Choice")

def main():
  movies = {
    "The Shawshank Redemption": 9.5,
    "Pulp Fiction": 8.8,
    "The Room": 3.6,
    "The Godfather": 9.2,
    "The Godfather: Part II": 9.0,
    "The Dark Knight": 9.0,
    "12 Angry Men": 8.9,
    "Everything Everywhere All At Once": 2.7,
    "Forrest Gump": 3.0,
    "Star Wars: Episode V": 8.7
    }

  outer = True
  print("----------Movie App---------")

  while outer:
    print("\n       1. List movies")
    print("       2. Add movies")
    print("       3. Delete movies")
    print("       4. Update movies")
    print("       5. Search movie")
    print("       6. Analytics")
    print("       7. Statistics")
    print("       8. Sort By Rating")
    print("       9. Random Movie")
    print("       0. Exit\n")

    choice = int(input("Enter your Choice: "))

    flag = True

    while flag:
      if(choice == 0):
        print("Successfully Terminated")
        outer = False
        flag = False

      elif(choice == 1):
        display(movies)
        flag = False

      elif(choice > 1 and choice < 6):
        movie = input("Enter movie name:")

        if(choice == 2):
          if(is_exist(movies, movie) is True):
            print("This movie already exists in the dictionary.")
          else:
            rating = float(input("Enter movie rating"))
            create(movies, movie, rating)
            flag = False

        elif(choice == 3):
          if(is_exist(movies, movie) is True):
            delete(movies, movie)
            flag = False
          else:
            print("This movie does not exit.")

        elif(choice == 4):
          print("You can only update the existing movie rating.")
          if(is_exist(movies, movie) is True):
            rating = float(input("Enter updating movie rating"))
            update(movies, movie, rating)
            flag = False
          else:
            print("This movie does not exit.")

        elif(choice == 5):
          if(is_exist(movies, movie) is True):
            read(movies, movie)
            flag = False
          else:
            print("This movie does not exit in the dictionary.")

        else:
          print("Invalid Choice")
          flag = False

      elif(choice == 6):
          analytics(movies)
          flag = False

      elif(choice == 7):
          statistics(movies)
          flag = False

      elif(choice == 8):
          sort(movies)
          flag = False

      elif(choice == 9):
          rand(movies)
          flag = False

      else:
        print("Invalid Choice")



main()
```
